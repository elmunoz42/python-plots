```python
# Streamlit app
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import streamlit as st
import matplotlib.pyplot as plt


st.title("SFS!")
input_data_url = st.text_input("Enter the URL to the dataset (csv file format). Ensure all variables are numbers")
if ".csv" not in input_data_url:
    st.write("No data")
    st.stop()

df = pd.read_csv(input_data_url)

col1, col2 = st.columns(2)

chosen_y = col1.selectbox("Choose a dependent variable:", df.columns)
if chosen_y == None:
    col1.write("Have one variable as the dependent")
    st.stop()

y = df.loc[:,chosen_y]
col1.write(y.head())

chosen_X = col2.multiselect("Choose independent variables:", df.drop(columns = chosen_y).columns)

X = df.loc[:,chosen_X]

col2.write(X.head())

if len(chosen_X) <= 1:
    col2.write("Please select more than one column")
    st.stop()

col1, col2 = st.columns(2)
k = col1.slider("Select the number of important features needed:", 1,
              X.shape[1], X.shape[1])

reg_or_class_option = col2.selectbox("Regression or Classification?", options=['regression', 'classification'])

lr, scoring = (LinearRegression(), 'neg_root_mean_squared_error') if reg_or_class_option == 'regression' else (LogisticRegression(), 'accuracy')
# lr, scoring = (LinearRegression(), 'neg_root_mean_squared_error')

sfs = SFS(lr,
          k_features=k,
          forward=True,
          scoring=scoring,
          cv=0)

sfs = sfs.fit(X, y)

sfs_metric = pd.DataFrame.from_dict(sfs.get_metric_dict()).T
st.write("Best features are:")
st.write(sfs_metric.iloc[-1,3])
st.line_chart(data = sfs_metric, y = 'avg_score')

st.write(sfs_metric)

```

```python
import pandas as pd

raw_df = pd.read_csv("https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv")
raw_df.head()
```

```python
raw_df.Class.unique()
```

```python
raw_df.Class.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X, y = raw_df.drop(columns = 'Class'), raw_df.Class

# split in to train test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# evaluate

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```

```python
y_pred_prob = model.predict_proba(X_test)
```

```python
y_pred_prob
```

```python
# extract the predicted probabilites for class 1 - the fraudulent class
probs = y_pred_prob[:,1]
```

```python
# define the thresholds
from numpy import arange
thresholds = arange(0,1,0.01)
```

```python
pd.Series([0.25,0.5,0.75]) >= 0.01
```

```python
# evaluate each threshold value
scores = [f1_score(y_test, (probs>=t).astype(int)) for t in thresholds]
```

```python
scores
```

```python
# get the best threshold
from numpy import argmax
ix = argmax(scores)
print(f"Threshold: {thresholds[ix]}, Best score: {scores[ix]}")
```

```python
y_pred_best_thresh = (probs>=thresholds[ix]).astype(int)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_best_thresh)
```

#OH15 - March 7, 2025

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function f(x) = x^2 * sin(x) + x
def cost_function(x):
    return x**1 * np.sin(x) + x

# Derivative of the cost function (gradient)
def gradient(x):
    return 1 * x * np.sin(x) + x**2 * np.cos(x) + 1


# Define the plot range for x
x_min, x_max = -6, 8  # These are the bounds for x_vals

# Generate x values for plotting the cost function
x_vals = np.linspace(x_min, x_max, 4000)
y_vals = cost_function(x_vals)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label='Cost Function', color='gray', linestyle='--')
```

```python
# Gradient Descent (GD)
def gradient_descent(starting_point, learning_rate, max_iterations):
    x = starting_point
    history = [x]

    for _ in range(max_iterations):
        grad = gradient(x)  # Compute the gradient
        x = x - learning_rate * grad  # Update the solution
        history.append(x)

    return x, history
```

```python
np.random.seed(123)
def simulated_annealing(starting_point, initial_temp, cooling_rate, max_iterations, x_min, x_max):
    x = starting_point        # Starting point is where we begin (like the initial temperature of the metal).
    temp = initial_temp       # Temperature controls how much the algorithm explores.
    history = [x]             # This keeps track of where we have been, kind of like recording the journey.

    for i in range(max_iterations):
        # Propose a new candidate solution by perturbing the current solution
        candidate = x + np.random.uniform(-1, 1)  # Small random changes in position (perturbation)
        candidate = np.clip(candidate, x_min, x_max)  # Clip candidate to make sure it's within the valid range (like controlling the boundaries).

        delta_cost = cost_function(candidate) - cost_function(x)  # Calculate how much the cost (energy) has changed.

        # If the new solution is better (lower cost), accept it; if not, accept it probabilistically.
        if delta_cost < 0 or np.random.rand() < np.exp(-delta_cost / temp):
            x = candidate  # Accept the candidate solution if it improves the result or based on probability.

        # Cool down the temperature gradually (reduce exploration and settle into the best solution).
        temp *= cooling_rate   # Cooling down (reducing temperature).
        history.append(x)      # Keep track of this solution in the history.

    return x, history
```

```python
# Set parameters
starting_point = 6  # Starting point for both algorithms
learning_rate = 0.01  # Lowered learning rate to avoid overflow
max_iterations = 5000
initial_temp = 10  # Initial temperature for SA
cooling_rate = 0.995  # Cooling rate for SA

# Define the plot range for x
x_min, x_max = -6, 8  # These are the bounds for x_vals

# Run Gradient Descent
gd_result, gd_history = gradient_descent(starting_point, learning_rate, max_iterations)

# Run Simulated Annealing with bounds for x
sa_result, sa_history = simulated_annealing(starting_point, initial_temp, cooling_rate, max_iterations, x_min, x_max)

# Generate x values for plotting the cost function
x_vals = np.linspace(x_min, x_max, 4000)
y_vals = cost_function(x_vals)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, label='Cost Function', color='gray', linestyle='--')
plt.scatter(gd_history, [cost_function(x) for x in gd_history], label='GD Path', color='blue', marker='o')
plt.scatter(sa_history, [cost_function(x) for x in sa_history], label='SA Path', color='red', marker='x')

# Highlight final results
plt.scatter(gd_result, cost_function(gd_result), color='black', zorder=5, label=f'GD Final: {gd_result:.2f}')
plt.scatter(sa_result, cost_function(sa_result), color='green', zorder=5, label=f'SA Final: {sa_result:.2f}')

# Labels and legend
plt.title('Comparison of Gradient Descent vs Simulated Annealing')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#########################################
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def f(x,y):
    "Objective function"
    return (x-3.14)**2 + (y-2.72)**2 + np.sin(3*x+1.41) + np.sin(4*y-1.73)

# Compute and plot the function in 3D within [0,5]x[0,5]
x, y = np.array(np.meshgrid(np.linspace(0,5,100), np.linspace(0,5,100)))
z = f(x, y)

# Find the global minimum
x_min = x.ravel()[z.argmin()]
y_min = y.ravel()[z.argmin()]

# The parameters c1 and c2 and are called the cognitive and the social coefficients respectively.
# We call the parameter the inertia weight constant. It is between 0 and 1 and determines how much should the particle keep on with its previous velocity (i.e., speed and direction of the search).
# Hyper-parameter of the algorithm
c1 = c2 = 0.1 # cognitive and social coefficients
w = 0.8 # inertia weight constant

# Create particles
n_particles = 20
np.random.seed(100)
X = np.random.rand(2, n_particles) * 5
V = np.random.randn(2, n_particles) * 0.1

# Initialize data
pbest = X
pbest_obj = f(X[0], X[1])
gbest = pbest[:, pbest_obj.argmin()]
gbest_obj = pbest_obj.min()

def update():
    "Function to do one iteration of particle swarm optimization"
    global V, X, pbest, pbest_obj, gbest, gbest_obj
    # Update params
    r1, r2 = np.random.rand(2)
    # Update particle velocities and positions
    V = w * V + c1*r1*(pbest - X) + c2*r2*(gbest.reshape(-1,1)-X)
    X = X + V
    # Update personal best positions and objective values
    obj = f(X[0], X[1])
    pbest[:, (pbest_obj >= obj)] = X[:, (pbest_obj >= obj)]
    pbest_obj = np.array([pbest_obj, obj]).min(axis=0)
    # Update global best position and objective value
    gbest = pbest[:, pbest_obj.argmin()]
    gbest_obj = pbest_obj.min()

print("PSO found best solution at f({})={}".format(gbest, gbest_obj))
print("Global optimal at f({})={}".format([x_min,y_min], f(x_min,y_min)))

# Set up base figure: The contour map
fig, ax = plt.subplots(figsize=(8,6))
fig.set_tight_layout(True)
img = ax.imshow(z, extent=[0, 5, 0, 5], origin='lower', cmap='viridis', alpha=0.5)
fig.colorbar(img, ax=ax)
ax.plot([x_min], [y_min], marker='x', markersize=5, color="white")
contours = ax.contour(x, y, z, 10, colors='black', alpha=0.4)
ax.clabel(contours, inline=True, fontsize=8, fmt="%.0f")
pbest_plot = ax.scatter(pbest[0], pbest[1], marker='o', color='black', alpha=0.5)
p_plot = ax.scatter(X[0], X[1], marker='o', color='blue', alpha=0.5)
p_arrow = ax.quiver(X[0], X[1], V[0], V[1], color='blue', width=0.005, angles='xy', scale_units='xy', scale=1)
gbest_plot = plt.scatter([gbest[0]], [gbest[1]], marker='*', s=100, color='black', alpha=0.4)
ax.set_xlim([0,5])
ax.set_ylim([0,5])

def animate(i):
    "Steps of PSO: algorithm update and show in plot"
    title = 'Iteration {:02d}'.format(i)
    # Update params
    update()
    # Set picture
    ax.set_title(title)
    pbest_plot.set_offsets(pbest.T)
    p_plot.set_offsets(X.T)
    p_arrow.set_offsets(X.T)
    p_arrow.set_UVC(V[0], V[1])
    gbest_plot.set_offsets(gbest.reshape(1,-1))
    return ax, pbest_plot, p_plot, p_arrow, gbest_plot

anim = FuncAnimation(fig, animate, frames=list(range(1,50)), interval=500, blit=False, repeat=True)
anim.save(filename="PSO.mp4", dpi=120)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/SavioSal/datasets/raw/master/diabetes.csv")
```

```python
df.head()
```

```python
X = df.drop(columns = "Outcome")
y = df.Outcome
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 123)
```

```python
# SMOTE - https://arxiv.org/pdf/1106.1813
```

```python
# dict to save the results
results = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1-score" : [],
    "Training Time": [],
    "Inference Time": []
}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Define model and train / evaluate
models = {
    'kNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC()
}
```

```python
import time
from sklearn import metrics

for model_name, model in models.items():
  start_time = time.time()
  model.fit(X_train, y_train)
  train_time = time.time() - start_time

  start_time = time.time()
  y_pred = model.predict(X_test)
  inference_time = time.time() - start_time

  accuracy = metrics.accuracy_score(y_test, y_pred)
  precision = metrics.precision_score(y_test, y_pred)
  recall = metrics.recall_score(y_test, y_pred)
  f1 = metrics.f1_score(y_test, y_pred)

  results['Model'].append(model_name)
  results['Accuracy'].append(accuracy)
  results['Precision'].append(precision)
  results['Recall'].append(recall)
  results['F1-score'].append(f1)

  results['Training Time'].append(train_time)
  results['Inference Time'].append(inference_time)

```

```python
pd.DataFrame(results)
```

```python
movies_new = movies[movies['vote_average']>7]
movies_new.shape
```

```python
movies_new = movies_new.reset_index()
```

```python
movies_new.tail()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies_new['overview'].isna().sum()
```

```python
movies_new['overview'].fillna("", inplace=True)
```

```python
movies_new['overview'].isna().sum()
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_new['overview'])
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
cosine_sim[0]
```

```python
indices = pd.Series(movies_new.index, index = movies_new['title']).drop_duplicates()
indices.head()
```

```python
similarity_scores = list(enumerate(cosine_sim[0]))
similarity_scores = sorted(similarity_scores, key = lambda x:x[1], reverse = True)
similarity_scores[1:11]
```

```python
def movie_recommendations(title):
  idx = indices[title]
  similarity_scores = list(enumerate(cosine_sim[idx]))
  similarity_scores = sorted(similarity_scores, key = lambda x:x[1], reverse = True)
  final_scores = similarity_scores[1:11]
  movie_indices = [i[0] for i in final_scores]
  return movies_new['title'].iloc[movie_indices]
```

```python
### collaborative filtering with association rules
transactions = [['apple', 'banana', 'orange'],
                ['banana', 'watermelon', 'grape'],
                ['grape', 'apple', 'banana'],
                ['banana', 'watermelon', 'orange'],
                ['apple', 'grape']]

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
user_history = ['apple', 'banana']
recommended_items = set()

for itemset in rules.antecedents:
    if set(itemset).issubset(user_history):
        for item in rules.loc[rules.antecedents == itemset, 'consequents']:
            recommended_items.update(item)


recommended_items
```
